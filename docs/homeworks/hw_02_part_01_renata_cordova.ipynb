## Visualización Distritos

```python
districts.head(1)
```

- **district_id**: El identificador único del distrito escolar
- **State**: El estado donde se ubica el distrito
- **locale**: Clasificación de la NCES que categoriza el territorio de los Estados Unidos en cuatro tipos: Ciudad, Suburbio, Pueblo y Rural.
- **pct_black/hispanic**: Porcentaje de estudiantes en los distritos identificados como Negros o Hispanicos. Basado en los datos de la NCES durante 2018-2019
- **pct_free/reduced** Porcentaje de estudiantes en los distritos que son candidatos para una beca alimenticia (especifícamente, para el almuerzo) parcial o total. Basado en los datos de la NCES durante 2018-2019
- **county_connections_ratio**: ratio (Conexiones de alta velocidad residencial fija sobre los 200 kilobits por segundo en al menos una dirección/hogar).
- **pp_total_raw**: gasto por pupilo (suma del gasto local y federal) de Edunomics Lab's National Education Resource Database on Schools project. La información del gasto es por cada escuela. Se usa el valor medio para representar el gasto para un distrito dado

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Distrubución de Estados', size = 20)
districts['state'].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
```

**Comentario:** Se observa que la mayor cantidad de personas estudiadas viven en Connecticut, Utah, Massachusetts e Illinois.

```python
fig, ax  = plt.subplots(figsize=(16, 8))
fig.suptitle('Distribución de Localidades', size = 20)
districts['locale'].value_counts().head(15).plot(kind = 'pie', autopct='%1.1f%%', figsize=(16, 8), startangle=0).legend();
plt.show()
```

**Comentario:** Vemos que la mayoría de de personas vive en suburbios o zonas rurales, respresentando un 59.1% y 18.8% de la población respectivamente. Este hecho sugiere que la mayoría de las personas participantes en el estudio podrían enfrentar limitaciones en el acceso a recursos educativos, como el internet.

```python
figure, axes = plt.subplots(figsize=(12, 5))

index = districts['pct_black/hispanic'].value_counts().index

graph = sns.countplot(
    data=districts,
    x= 'pct_black/hispanic',
    order=index,
    ax=axes
)

total = len(districts['pct_black/hispanic'])
total_nulos = len(districts['pct_black/hispanic'][districts['pct_black/hispanic'].isnull()])
for p in graph.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2 - 0.1
    y = p.get_height()
    graph.annotate(percentage, (x, y), ha='center', va='bottom')

axes.set_title('Black/hispanic')
plt.show()

print(f'Porcentaje de datos nulos: {(100*total_nulos/total)}%' )
```

**pct_black/hispanic:** Porcentaje de estudiantes en los distritos identificados como Negros o Hispanicos. Basado en los datos de la NCES durante 2018-2019

**Comentario:** Vemos que la mitad de los distritos posee una identificación con negro/hispano de entre un 0 a 20%, mientras que solo un 4.5% de los distritos posee una identificación de entre 80% a 100%. También, que a medida que aumenta el porcentaje de identificación, más bajo es el porcentaje de la población representada.

**Comentarios:**

El promedio diario de productos accedidos está alrededor de 300-500 por distrito escolar en 2020.

Cada mes presenta fluctuaciones, que sigue un patrón de 5-2, es decir, 5 días de escuela y 2 días de fin de semana. Aunque el número de productos accedidos disminuye durante el fin de semana, aún observamos productos accedidos en esos días, siendo aproximadamente un 25%-50% en comparación con los días de la semana. Esto tiene sentido, ya que los fines de semana no son días de clases.

A mediados de febrero, en el cierre temporal de escuelas por la pandemia, vemos un aumento en los productos accedidos, aunque el aumento es marginal. Luego, observamos una baja durante las vacaciones de verano y, después, un aumento en el promedio diario de productos accedidos.
